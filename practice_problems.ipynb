```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data

# Perform PCA to reduce dimensionality
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Create a KMeans model with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)

# Fit the model to the reduced data
kmeans.fit(X_pca)

# Get the cluster centers and labels for each data point
centers = kmeans.cluster_centers_
labels = kmeans.labels_

# Create a scatter plot for the data points with color-coded clusters
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis', s=50, alpha=0.5, label='Data Points')

# Highlight the cluster centroids in red
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.8, label='Cluster Centroids')

# Set plot title and axis labels
plt.title('K-means Clustering on Iris Dataset (PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

# Display legend
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
import skfuzzy as fuzz
from sklearn.decomposition import PCA

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data

# Perform PCA to reduce dimensionality
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Apply fuzzy c-means clustering
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X)
cntr, u, _, _, _, _, _ = fuzz.cluster.cmeans(
    X_pca.T, c=3, m=2, error=0.005, maxiter=1000, init=None)
centers_pca = pca.inverse_transform(cntr.T)
# Transpose the cluster centers to match the original data shape
centers_pca = pca.inverse_transform(cntr.T)

# Create a scatter plot for the data points with color-coded clusters
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=np.argmax(u, axis=0), cmap='viridis', s=50, alpha=0.5, label='Data Points')

# Highlight the cluster centroids in red
plt.scatter(centers_pca[:, 0], centers_pca[:, 1], c='red', s=200, alpha=0.8, label='Cluster Centroids')

# Set plot title and axis labels
plt.title('Fuzzy C-means Clustering on Iris Dataset (PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

# Display legend
plt.legend()

# Show the plot
plt.show()

print(cntr.shape)
print(pca.components_.shape)
```
