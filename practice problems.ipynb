```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data

# Create a KMeans model with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)

# Fit the model to the data
kmeans.fit(X)

# Get the cluster centers and labels for each data point
centers = kmeans.cluster_centers_
labels = kmeans.labels_

# Create a scatter plot for the data points with color-coded clusters
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, alpha=0.5, label='Data Points')

# Highlight the cluster centroids in red
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.8, label='Cluster Centroids')

# Set plot title and axis labels
plt.title('K-means Clustering on Iris Dataset')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

# Display legend
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
import skfuzzy as fuzz

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data

# Apply fuzzy c-means clustering
cntr, u, u0, d, jm, p, fpc = fuzz.cluster.cmeans(
    X.T, c=3, m=2, error=0.005, maxiter=1000, init=None)

# Transpose the cluster centers to match the original data shape
centers = cntr.T

# Assign data points to the cluster with the highest membership
labels = np.argmax(u, axis=0)

# Create a scatter plot for the data points with color-coded clusters
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, alpha=0.5, label='Data Points')

# Highlight the cluster centroids in red
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.8, label='Cluster Centroids')

# Set plot title and axis labels
plt.title('Fuzzy C-means Clustering on Iris Dataset')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

# Display legend
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn_extra.cluster import KMedoids

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data

# Create a KMedoids model with 3 clusters
kmedoids = KMedoids(n_clusters=3, random_state=0)

# Fit the model to the data
kmedoids.fit(X)

# Get cluster labels and medoids (cluster centroids)
labels = kmedoids.labels_
medoids = kmedoids.cluster_centers_

# Create a scatter plot for the data points with color-coded clusters
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, alpha=0.5, label='Data Points')

# Highlight the cluster medoids in red
plt.scatter(medoids[:, 0], medoids[:, 1], c='red', s=200, alpha=0.8, label='Cluster Medoids')

# Set plot title and axis labels
plt.title('K-medoids Clustering on Iris Dataset')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

# Display legend
plt.legend()

# Show the plot
plt.show()
```
